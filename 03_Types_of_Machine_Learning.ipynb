# 📌 Introduction

Machine Learning (**ML**) is a branch of Artificial Intelligence (**AI**) that enables computers to learn patterns from data and make predictions or decisions without being explicitly programmed.

However, not all Machine Learning models learn in the same way. Depending on the type of data available and the learning objective, Machine Learning is divided into **three major categories**:

- 🎯 **Supervised Learning**
- 🔍 **Unsupervised Learning**
- 🎮 **Reinforcement Learning**

And exclusively there is another category:

- 💻 **Semi-Supervised Learning**

Each type is designed to solve different kinds of problems and is widely used in real-world AI applications, from spam detection and recommendation systems to self-driving cars and robotics.

This notebook explains how each learning approach works, their key differences, practical examples, and where they are used in the real world.

# 🎯 Supervised Learning

### *Learning from Labeled Data*

Supervised Learning is the most common type of Machine Learning. In this approach, the model is trained using **labeled data**, meaning every input already has the correct output (answer).

The goal of the model is to learn the relationship between inputs and outputs so that it can accurately predict the output for new, unseen data.

## 💡 Simple Idea

Think of a teacher solving questions with answers written beside them.

The student (ML model) studies many solved examples and eventually learns how to solve new questions on its own.

**Example**

| Input | Output |
|-------|--------|
| House Features | House Price |
| Student Study Hours | Exam Score |
| Email | Spam / Not Spam |
| Medical Data | Disease Prediction |

## ⚙️ How It Works

1. Collect labeled data.
2. Train the machine learning model.
3. Learn patterns between input and output.
4. Predict outputs for new data.

## 📌 Real-World Applications

- 📧 Spam Email Detection
- 🏠 House Price Prediction
- 💳 Credit Card Fraud Detection
- 🩺 Disease Diagnosis
- 😊 Sentiment Analysis
- 🚗 Traffic Sign Recognition

## 🧠 Types of Supervised Learning

There are **two major** tasks:

### 📈 1. Regression

Used when the output is a **continuous numerical value**.

**Examples**
- Predicting house prices
- Predicting temperature
- Predicting sales


### 🏷️ 2. Classification

Used when the output belongs to a **category or class**.

**Examples**
- Spam or Not Spam
- Cat or Dog
- Fraud or Genuine Transaction
- Positive or Negative Review


## ✅ Key Characteristics

- Uses **labeled data**
- Learns from correct answers
- High prediction accuracy when quality data is available
- Most widely used Machine Learning approach
- Suitable for prediction and classification problems
